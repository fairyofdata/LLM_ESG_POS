```python
from bertopic import BERTopic
from transformers import AutoModel, AutoTokenizer

# KOBERT model and torque knight rod (model name is an example. You need to check the actual model name.)
model_name = "monologg/kobert"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

Applying the kobert model to # bertopic
topic_model = BERTopic(embedding_model=(model, tokenizer))
```

## Remove tokenization and insoluble language

```python
!pip install konlpy
```

```python
from konlpy.tag import Okt
import pandas as pd

# OKT Talk Nizor initialization
okt = Okt()

# Instrument list load function
def load_stopwords(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        stopwords = [line.strip() for line in file.readlines()]
    return stopwords

stopwords_path = '/content/drive/MyDrive/Kwargs/stopwords.txt'  # 불용어 리스트 파일 경로
stopwords = load_stopwords(stopwords_path)

# In the text (OKT version) in text (OKT version)
def remove_stopwords_with_okt(text, stopwords):
# Tokenization using OKT
    tokens = okt.morphs(text)
# Remove insoluble language
    filtered_tokens = [token for token in tokens if token not in stopwords]
# Together again with strings
    filtered_text = ' '.join(filtered_tokens)
    return filtered_text

# Uninstalled removal for each document in the dataset
filtered_doclist = [remove_stopwords_with_okt(doc, stopwords) for doc in df['content']]

# Output
for doc in filtered_doclist:
    print(doc)
```

## bertopic

```python
#Bertopic model training
doclist = df['content'].tolist()
topics, probs = topic_model.fit_transform(doclist) # 토큰화 결과 : doc
#Topular output
topic_info = topic_model.get_topic_info()
print(topic_info)

#Plain word output for specific topics
topic_number = 0  # 예시로 0번 주제를 선택
topic = topic_model.get_topic(topic_number)
print(f"Topic {topic_number} Representative Words and their Scores:")
for word, score in topic:
    print(f"{word}: {score}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df = topic_info
# Create a new 'TOPIC_NAME' column by combining the subject number and the theme name.
df['Topic_Name'] = df['Topic'].astype(str) + ": " + df['Name'].apply(lambda x: x.split('_')[1])
```
