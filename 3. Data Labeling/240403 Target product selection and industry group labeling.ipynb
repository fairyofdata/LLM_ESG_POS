### Calculation of weighted average order

```python
import pandas as pd

df7 = df4

#Set weight setting
시총순위_가중치 = 0.5
기사발생량순위_가중치 = 0.5

# Weighted average calculation
df7['가중평균순위'] = (df7['산업군내시총순위'] * 시총순위_가중치 +
                 df7['산업군내기사발생량순위'] * 기사발생량순위_가중치)

# DataFrame alignment according to the new ranking (optional)
df7 = df7.sort_values('가중평균순위')

df7
```

```python
import pandas as pd

df = df7

#Set weight setting
시총순위_가중치 = 0.5
기사발생량순위_가중치 = 0.5

# Weighted average ranking calculation
df['가중평균순위'] = (df['산업군내시총순위'] * 시총순위_가중치 +
                 df['산업군내기사발생량순위'] * 기사발생량순위_가중치)

# DataFrame alignment according to the new ranking
df_sorted = df.sort_values('가중평균순위').reset_index(drop=True)

# Calculation between weighted average and existing rankings
df_sorted['시총순위대비값'] = df_sorted['가중평균순위'] - df_sorted['산업군내시총순위']
df_sorted['기사발생량순위대비값'] = df_sorted['가중평균순위'] - df_sorted['산업군내기사발생량순위']

df_sorted
```

```python
# Calculation of the gap rate of the price ranking and the amount of the price of the article
df_sorted['괴리율'] = abs(df_sorted['시총순위대비값'] - df_sorted['기사발생량순위대비값'])

#Sorted in the order of companies with the largest gap rate
df_sorted_by_discrepancy = df_sorted.sort_values('괴리율', ascending=False)

#Please output and numerical output with the largest gap rate
df_sorted_by_discrepancy.head(30)
```

```python
# Attempting to convert the 'Knights' column to an integer type
df_sorted['기사발생량'] = pd.to_numeric(df_sorted['기사발생량'], errors='coerce')

# Reconfirmation of data type after conversion
df_sorted['기사발생량'].dtype
```

```python
# Initial setting according to the number of companies already assigned
initial_companies_assigned = {'나연': 7, '지헌': 5, '한재': 2, '해리': 1}

# Dictionary initialization to track the collecting articles allocated by person in charge
collector_totals = {'나연': 50000, '지헌': 40000, '한재': 5000, '해리': 0000}
```

```python
# Separation of companies that are already allocated and unexplained companies
assigned_df = df_sorted.dropna(subset=['완료'])
unassigned_df = df_sorted[df_sorted['완료'].isna()]

# Initialization of dictionary that tracks the number of companies allocated by person in charge in consideration of the number of companies already assigned.
company_totals = initial_companies_assigned.copy()

# Update for the number of companies already allocated
for collector in ['나연','지헌','한재','해리']:
    company_totals[collector] = assigned_df[assigned_df['완료'] == collector].shape[0]

# Application of distribution logic for the remaining companies
collectors_updated = []

for _, row in unassigned_df.iterrows():
# Condition 1: Choosing a person in charge of the least company
# Condition 2: If you are in charge of the same number, select the person in charge with the least collecting articles.
    min_collector = min(company_totals, key=lambda x: (company_totals[x], collector_totals[x]))

# Allocate a company to the person in charge
    collectors_updated.append(min_collector)

# Update the total number of companies and the amount of art collection of allocated staff
    company_totals[min_collector] += 1
    collector_totals[min_collector] += row['기사발생량']

# Added the assigned personnel information to the company's data frame.
unassigned_df['완료'] = collectors_updated

# All -time corporate data frame and already assigned corporate data frame combined
final_df = pd.concat([assigned_df, unassigned_df]).sort_index()

# Confirmation of the number of business number and article collection by the final allocated person in charge

final_company_totals = pd.Series(company_totals)
print('할당종목수')
print(final_company_totals)

final_article_totals = pd.Series(collector_totals)
print('할당기사수(이미 한 거 제외)')
print(final_article_totals)
```
